```python
def countWords(words, words_list):
    words_count = []
    for word in words_list:
        words_count.append(words.count(word))
    return words_count

def eraseStop(txt):
    nouns = []
    for noun in txt:
        if noun not in stopwords:
            nouns.append(noun)
    return nouns

def eraseOne(nouns):
    words = []
    for noun in nouns:
        if len(noun) > 1:
            words.append(noun)
    return words
```
